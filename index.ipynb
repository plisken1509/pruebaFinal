```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_folder = "CarneDataset/train/"
test_folder = "CarneDataset/test/"
```

```python
train_images = []

# Recorre las imágenes en la carpeta de entrenamiento
for filename in os.listdir(train_folder):
    img = cv2.imread(os.path.join(train_folder, filename))
    if img is not None:
        train_images.append(img)
```

```python
test_images = []

# Recorre las imágenes en la carpeta de prueba
for filename in os.listdir(test_folder):
    img = cv2.imread(os.path.join(test_folder, filename))
    if img is not None:
        test_images.append(img)
```

```python
num_images_to_show = 6  # Número de imágenes a mostrar
fig, axs = plt.subplots(2, 3, figsize=(10, 6))
for i, ax in enumerate(axs.flatten()):
    if i < len(train_images):
        ax.imshow(cv2.cvtColor(train_images[i], cv2.COLOR_BGR2RGB))
        ax.axis('off')
plt.show()
```
